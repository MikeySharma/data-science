```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interact, widgets

%matplotlib inline

# Load the iris dataset
iris = sns.load_dataset('iris')
print(iris.head())
# Create an interactive scatter plot using ipywidgets
def plot_scatter(x_axis, y_axis, hue_by):
    sns.scatterplot(x=x_axis, y=y_axis, hue=hue_by, data=iris)
    plt.show()

interact(plot_scatter,
         x_axis=widgets.Dropdown(options=iris.columns, value='sepal_length'),
         y_axis=widgets.Dropdown(options=iris.columns, value='sepal_width'),
         hue_by=widgets.Dropdown(options=['species', None], value='species'))
```

```python
# Create an interactive scatter plot using ipywidget
df = pd.read_csv('../data-sets/mtcars.csv')

%matplotlib inline

def plot_scatter(x_axis, y_axis, hue_by):
    sns.scatterplot(x=x_axis, y=y_axis, hue=hue_by, data=df)
    plt.show()

interact(plot_scatter,
         x_axis=widgets.Dropdown(options=df.columns, value='cyl'),
         y_axis=widgets.Dropdown(options=df.columns, value='qsec'),
         hue_by=widgets.Dropdown(options=['carb', None], value='carb'))
```
