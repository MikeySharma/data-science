```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score, confusion_matrix, recall_score, precision_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../data-sets/fitness-data.csv')
df.info()
```

```python
df.set_index('epoch (ms)', inplace=True)
df.head()
```

```python
def map_to_replace(columns):
    mapValue = {}
    i = 0
    for col in columns:
        if df[col].dtype == 'object':
            for unique_val in df[col].unique():
                mapValue[str(unique_val)] = i
                i+=1
            df[col] = df[col].map(mapValue)
            mapValue = {}
            i = 0

map_to_replace(['Label', 'Category', 'Participants'])
```
