```python
#import necessary depenencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score, f1_score
from sklearn.model_selection import train_test_split
```

```python
#import data csv file
df = pd.read_csv('../data-sets/Housing-1.csv')
print(df.head())
```

```python
df['furnishingstatus'].value_counts()
```

```python
def replace_with_mapping(columns, mapping_value):
    for col in columns:
        df[col]= df[col].map(mapping_value)

values = {
    'yes' : 1,
    'no' : 0
}
values2 = {
    'furnished' : 2,
    'semi-furnished': 1,
    'unfurnished' : 0
}
```

```python
#replace the string data to float or int data
df = df.dropna()
# df.loc[:, ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']] = df.loc[:, ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']].replace(['yes', 'no'], [1, 0])
# df.loc[:, 'furnishingstatus'].replace(['furnished', 'semi-furnished', 'unfurnished'], [2, 1, 0], inplace=True)
replace_with_mapping(np.array(['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']), values)
replace_with_mapping(np.array(['furnishingstatus']), values2)
```

```python
# print(df['bathrooms'].value_counts)
df.corr()
```

```python
#Split data into X and y
X = df.drop(['price', 'basement', 'hotwaterheating', 'guestroom'], axis=1)
y = df['price']

threshold = y.median()
y_binary = (y >= threshold).astype(int)
```

```python
#Split data into train and test data
x_train, x_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
#initialize and train model
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

```python
#predict the data
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
confusionMatrx = confusion_matrix(y_test, y_pred)
reclScore = recall_score(y_test, y_pred)
precisionScore = precision_score(y_test, y_pred)
f1Score = f1_score(y_test, y_pred)

print(f'Accuracy : {accuracy}')
print(f'Confusion Matrix : {confusionMatrx}')
print(f'Recall Score: {reclScore}')
print(f'Precision Score: {precisionScore}')
print(f'F1 Score : {f1Score}')
```
